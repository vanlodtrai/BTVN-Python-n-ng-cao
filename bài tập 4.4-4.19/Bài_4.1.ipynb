```python
import pandas as pd
import numpy as np

# Tạo chuỗi pandas từ danh sách
mylist = list('abcdefghijklmnopqrstuvwxyz')
ser_from_list = pd.Series(mylist)
print("Series từ danh sách:")
print(ser_from_list)

# Tạo chuỗi pandas từ numpy array
myarr = np.arange(26)
ser_from_array = pd.Series(myarr)
print("\nSeries từ numpy array:")
print(ser_from_array)

# Tạo chuỗi pandas từ từ điển
mydict = dict(zip(mylist, myarr))
ser_from_dict = pd.Series(mydict)
print("\nSeries từ từ điển:")
print(ser_from_dict)
```
