```python
import pandas as pd
```

```python
# Đọc file euro12.csv
euro12 = pd.read_csv("C:/Users/User/OneDrive/Tài liệu/GitHub/12_Duc_17A2/BTVN_Chương_4/euro2012.csv")

# 1. In giá trị cột Goals
print("Cột Goals:")
print(euro12["Goals"])

# 2. Có bao nhiêu đội tham gia Euro2012?
print("\nSố đội tham gia Euro2012:", euro12["Team"].nunique())

# 3. In thông tin của Euro2012
print("\nThông tin Euro2012:")
print(euro12.info())

# 4. Tạo DataFrame mới có 3 cột: Team, Yellow Cards, Red Cards
discipline = euro12[["Team", "Yellow Cards", "Red Cards"]]
print("\nDataFrame mới - Discipline:")
print(discipline)

# 5. Sắp xếp giảm dần theo Red Cards và Yellow Cards
discipline_sorted = discipline.sort_values(by=["Red Cards", "Yellow Cards"], ascending=False)
print("\nSắp xếp giảm dần theo Red Cards và Yellow Cards:")
print(discipline_sorted)

# 6a. Trung bình Yellow Cards
print("\nTrung bình số Yellow Cards:", discipline["Yellow Cards"].mean())

# 6b. Lọc các đội ghi hơn 6 bàn thắng
teams_more_than_6_goals = euro12[euro12["Goals"] > 6]
print("\nCác đội ghi hơn 6 bàn thắng:")
print(teams_more_than_6_goals)

# 7. Các đội bắt đầu bằng chữ "G"
teams_start_with_G = euro12[euro12["Team"].str.startswith("G")]
print("\nCác đội bắt đầu bằng chữ G:")
print(teams_start_with_G)

# 8. In 7 cột đầu của DataFrame euro12
print("\n7 cột đầu của DataFrame:")
print(euro12.iloc[:, :7])

# 9. In tất cả các cột, trừ 3 cột cuối
print("\nTất cả các cột trừ 3 cột cuối:")
print(euro12.iloc[:, :-3])

# 10. In các cột: Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
selected_columns = euro12[["Team", "Goals", "Shooting Accuracy", "Yellow Cards", "Red Cards"]]
print("\nCác cột đã chọn:")
print(selected_columns)

# 11. Chỉ hiện thị Team, Shooting Accuracy từ các đội: England, Italy, Russia
filtered_teams = euro12[euro12["Team"].isin(["England", "Italy", "Russia"])]
print("\nThông tin của các đội England, Italy, Russia:")
print(filtered_teams[["Team", "Shooting Accuracy"]])
```
